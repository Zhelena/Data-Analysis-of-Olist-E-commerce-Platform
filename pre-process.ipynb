```python
import numpy as np 
import pandas as pd
```

```python
customers = pd.read_csv('olist_customers_dataset.csv')
geolocation = pd.read_csv('olist_geolocation_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
category = pd.read_csv('product_category_name_translation.csv')
```

```python
print('reviews: \n',reviews.isnull().sum(),'\n')
print('orders: \n',orders.isnull().sum(),'\n')


print('customers: \n',customers.isnull().sum(),'\n')
print('geolocation: \n',geolocation.isnull().sum(),'\n')
print('items: \n',items.isnull().sum(),'\n')
print('payments: \n',payments.isnull().sum(),'\n')
print('category: \n',category.isnull().sum())
```

```python
orders = orders.dropna()
orders.to_csv('orders.csv', index=False)
```

```python
# customer_id 是否有重复值
customer_id_duplicates = customers['customer_id'].duplicated().any()

# customer_unique_id 是否有重复值
customer_unique_id_duplicates = customers['customer_unique_id'].duplicated().any()

if customer_id_duplicates:
    print("customer_id 列存在重复值")
else:
    print("customer_id 列没有重复值")

if customer_unique_id_duplicates:
    print("customer_unique_id 列存在重复值")
else:
    print("customer_unique_id 列没有重复值")
```

```python
customers.groupby('customer_unique_id').filter(lambda x: len(x) > 1).sort_values(by='customer_unique_id')
```

```python
# 合并两个 DataFrame，根据相同的列名进行合并
merged_df = pd.merge(customers, geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='inner')
merged_df
```

```python
# 把order中字符串转化为日期时间数据类型
def transform_datetime(data, colum_list):
    for i in colum_list:
        data[i] = pd.to_datetime(data[i])
    print('数据类型转化成功！\n')
    
colum_list = ['order_purchase_timestamp', 'order_approved_at',
              'order_delivered_carrier_date', 'order_delivered_customer_date', 
              'order_estimated_delivery_date']
transform_datetime(orders,colum_list)   
orders.info()
```
