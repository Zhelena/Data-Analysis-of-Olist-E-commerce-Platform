```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
payments = pd.read_csv('olist_order_payments_dataset.csv')
```

```python

payment_counts = payments['payment_type'].value_counts()

# 
colors = ['skyblue', 'lightgreen', 'pink', 'orange', 'purple']

# 
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.4), colors=colors)
plt.title('Payment Types')

# 
plt.legend(wedges, payment_counts.index, title='Payment Types', loc='center left', bbox_to_anchor=(1, 0, 0.5, 1))

plt.axis('equal')  # 
plt.show()
```

```python
# 
installments_percentage = payments['payment_installments'].value_counts(normalize=True) * 100

# 
plt.figure(figsize=(12, 6))
bars = plt.bar(installments_percentage.index, installments_percentage, color='skyblue', label='Payment Installments')

# 
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{round(yval, 2)}%', ha='center', va='bottom')

# 
plt.title('Payment Installments')
plt.xlabel('Number of Installments')
plt.ylabel('Percentage')

# 
plt.legend()

# 
plt.show()
```
